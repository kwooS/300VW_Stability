```python
from google.colab import drive
import numpy as np
import os
import cv2

drive.mount('/content/gdrive/')
```

```python
#change video into frames
def video_to_frames(dir, name): #possible names: "test1", "test2", "test3", "training"
  frames_path_list = []
  annot_path_list = []
  for i in os.listdir(dir): #goes over test or training
    print("looking over main dir")
    print(i)
    if i == name:
      print("entering file "+i)
      file1 = os.path.join(dir, i) #training, or test1, 2, 3
      if os.path.isdir(file1): #i == name: #원하는 name을 가진 파일만 확인
        for j in sorted(os.listdir(file1)): #goes over number files eg 001, 002
          file2 = os.path.join(file1, j) # 001, 002 etc
          print("starting at " + file2) 
          annot_path = os.path.join(file2, "annot")
          frames_path = os.path.join(file2, "frames")
          video_path = os.path.join(file2, "vid.avi")          
          if not os.path.exists(frames_path):
            print("creating folder frames")
            os.mkdir(frames_path) #create frames folder

            vidcap = cv2.VideoCapture(video_path)
            success,image = vidcap.read()
            count = 0

            os.chdir(frames_path) #change directory to frames_path to save frames there
            print("adding frames to file...")
            while success:
              cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
              success,image = vidcap.read()
              count += 1   
            frames_path_list.append(frames_path)
            annot_path_list.append(annot_path)
            print("completed adding frames")
  return frames_path_list, annot_path_list       
```

```python
#change training videos into frames
root = "/content/gdrive/My Drive/ESTsoft Internship/300VW_Dataset_2015_12_14"

training_frames_path, training_annot_path = video_to_frames(root, "training")
```
